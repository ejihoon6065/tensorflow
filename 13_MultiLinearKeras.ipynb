```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 케라스는 세가지 임포트
# 내부적으로 tensor로 바꾸는 기능이 존재

from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam 

# 데이터 셋 불러오기 위해 추가적 임포트
from sklearn import datasets
```

```python
iris = datasets.load_iris()


# 2차원, 1번째 열을 레이블, 나머지 열을 feature로 사용
data = iris['data']
```

```python
# Dense에서 입력의 개수 출력의 개수 조정
xdata = data[:, 1:]
ydata = data[:, [0]]
ydata
```

```python
#  feature의 개수 3개
IO = Dense(units=1, input_dim = 3)

# 순차 모델 ; 쌓아올리기
# Sequential - 대문자 
model = Sequential( [IO] )
model.compile( loss = 'mean_squared_error', optimizer = Adam(0.1))
h = model.fit( xdata, ydata, epochs = 1000)
```

```python
plt.plot( h.history['loss'] )

#loss 값이 줄어드는 모습이 나타남
plt.show
```

```python
# 3행 1렬로 나타남
# 내부적으로 matmul

IO.get_weights()
```

```python
# numpy 형태로 예측해야 함

#model.predict([3.5,1.4,0.2]) -> 오류발생
model.predict(np.array([[3.5,1.4,0.2]]))
```

```python
plt.plot(ydata)
plt.plot(model.predict(xdata), 'r--')
plt.show()
```
