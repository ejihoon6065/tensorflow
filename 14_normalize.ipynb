```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 정규화
from sklearn import preprocessing   # 전처리
```

# 방법 1

```python
# 0~ 3번지 컬럼 / 4번지 컬럼
# 0~3번째까지 feature
# 맨 마지막은 레이블
 
data = [[828, 920, 1234567, 1020, 1111],

        [824, 910, 2345612, 1090, 1234],

        [880, 900, 3456123, 1010, 1000],

        [870, 990, 2312123, 1001, 1122],

        [860, 980, 3223123, 1008, 1133],

        [850, 970, 2432123, 1100, 1221]]

data = np.float32(data)
```

```python
# 0:4 
# 0<= column <4 
# 0:-1
# 0 <= column <-1

xdata = data[:,:-1]
ydata = data[:,[-1]]
```

```python
ydata
```

```python
# feature의 개수가 4개

x = tf.placeholder( tf.float32 , shape =[None,4]) #---------> n행 3열 행렬

y = tf.constant( ydata, tf.float32)
```

```python
# feature의 개수 4개 

w = tf.Variable( tf.random_uniform([4,1]))
b = tf.Variable( tf.random_uniform([1]))
```

```python
hx = tf.matmul( x,w ) + b    # n x 4 행렬   +      4 x 1 행렬    => n x 1 행렬

cost = tf.reduce_mean(tf.square(hx-y))
```

```python
# argument로 learning mate를 투입, 정규화를 위해 매우 작은 값 투입
optimizer = tf.train.AdamOptimizer( 0.0001 ) 

train = optimizer.minimize( cost )
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
# 트레이닝
for i in range(10000) :
    # x가 placeholder이므로 feed_dict 줘야 함
    sess.run( train, {x:xdata} )
    print(i, sess.run(cost, {x:xdata}))
```

```python
# 결과 Nan에 대한 해결책

# 1. 러닝메이트 값을 매우 작게 주는 방법   -> 비 추천
# 2. optimizer를 바꾸는 방법 - Adam 사용  -> 추천
# 3. normalize (정규화)                  -> 가장 추천

# normalization
# (X - Xmin) / (Xmax - Xmin)


# standardization
# 표준화
# (X - 평균) / 표준편차
```

# 방법 2

```python
scale = preprocessing.MinMaxScaler()
dataN = scale.fit_transform( data )
```

```python
# 0에서 1까지의 값으로 변경됨
# normalization 수행
dataN
```

```python
xdata =  dataN[:,:-1]
ydata = dataN[:,[-1]]

# feature의 개수가 4개

x = tf.placeholder( tf.float32 , shape =[None,4]) #---------> n행 4열 행렬
y = tf.constant( ydata, tf.float32)

w = tf.Variable( tf.random_uniform([4,1]))
b = tf.Variable( tf.random_uniform([1]))

hx = tf.matmul( x,w ) + b    # n x 4 행렬   +      4 x 1 행렬    => n x 1 행렬
cost = tf.reduce_mean(tf.square(hx-y))
```

```python
optimizer = tf.train.GradientDescentOptimizer( 0.00001 ) 

train = optimizer.minimize( cost )

sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

for i in range(10000) :
    # x가 placeholder이므로 feed_dict 줘야 함
    sess.run( train, {x:xdata} )
    print(i, sess.run(cost, {x:xdata}))
```

```python
# weight
sess.run(w)
```

```python
# bias
sess.run(b)
```
