```python
#MATMUL

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# hearder의 디폴트가 첫번째 값으로 되어 있으므로 None으로 해제시킨다-> 데이터로 빠짐

df = pd.read_csv('data-01.csv', header = None)
df.columns = ['q1', 'q2', 'mid', 'score']
df
```

```python
# 예측값을 'placeholder'로 잡아야 하기 때문에 넣어줌

xdata = df[['q1', 'q2', 'mid']] #--------------> n행 3열 행렬 


# y = df['score']    -------------> y값이 '시리즈'로 생성됨

# 따라서 y값을 '행렬'로 만들기 위해
ydata = df[['score']]
```

```python
# 예측을 위해 'placeholder'로 만듦

x = tf.placeholder( tf.float32 , shape =[None,3]) #---------> n행 3열 행렬

y = tf.constant( ydata, tf.float32)
```

```python
# [3,1]  ->  한마디로 w는 [피쳐의 개수, 라벨의 개수]가 된다
w = tf.Variable( tf.random_uniform([3,1]))
b = tf.Variable( tf.random_uniform([1]))
```

```python
hx = tf.matmul( x,w ) + b    # n x 3 행렬   +      3 x 1 행렬    => n x 1 행렬

cost = tf.reduce_mean(tf.square(hx-y))
```

```python
# argument로 learning mate를 투입, 정규화를 위해 매우 작은 값 투입
optimizer = tf.train.GradientDescentOptimizer( 0.00001 ) 

train = optimizer.minimize( cost )
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
for i in range(10000) :
    # x가 placeholder이므로 feed_dict 줘야 함
    sess.run( train, {x:xdata} )
    print(i, sess.run(cost, {x:xdata}))
```

```python
# w1, w2, w3 출력
sess.run(w)
```

```python
# q1이 73점, q2가 80점 ,mid가 75점
# 행렬로 예측

sess.run(hx, {x:[[73,80,75]]})
```
