```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost (x,y,w) :
    c = 0
    for i in np.arange( len(x) ) :
        hx = w*x[i]
        c = c + (hx -y[i]) ** 2
    return c/len(x)
```

```python
def gradient (x,y,w) : # 기울기...
    c = 0
    for i in np.arange( len(x) ) :
        hx = w*x[i]
        c = c + (hx -y[i]) *x[i]  # w로 편미분 한 후의 결과 식을 만든 것
    return c/len(x)
```

```python
def show_gradient():
    x = [1,2,3]
    y = [1,2,3]
    w = 10                 # w값은 임의의 값 선정한 후 시작.
    for i in range(200) :  # 루프도 충분히 돌려야 함. 학습 완료되면 빠져나가는 방법도 존재.
        c = cost(x,y,w)
        g = gradient(x,y,w) # 기울기
        w = w-0.1*g         # 0.1(running mate)은 원하는 값 대입(보폭). 작을 수록 최저값 유추하기 유리. 
        print(i,c)
    print('w:' ,w)
 
show_gradient()
```
