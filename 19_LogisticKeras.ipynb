```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 세가지
# 케라스는 텐서객체를 반드시 줄 필요 없음. 시리즈 등등 가능
# 내부적으로 tensor로 바꾸는 기능이 있음

from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam 
```

```python
df = pd.read_csv('data/diabetes1.csv', header = None)
df
```

```python
xdata = df.iloc[:, :-1] 
ydata = df.iloc[:,[-1]]
```

```python
IO = Dense( units = 1, input_dim = 8, activation = 'sigmoid') # 입력값 8개
model = Sequential( [IO] )
model.compile(loss = 'binary_crossentropy', 
              optimizer = Adam(0.1), metrics=['accuracy']) #로그함수 사용
h = model.fit(xdata,ydata, epochs=1000) # 1000번 돌림
```

```python
# 예측값 : 시그모이드 함수

model.predict(xdata)
```

```python
# 캐스팅까지 정의 : 0에서 1로 나옴
# classes : 'tf.cast(hx>0.5, dtype = tf.float32)'가 정의되어 있음
model.predict_classes(xdata)
```

```python
h.history['acc'][-1]
# 제일 마지막
```

#### 결론 : Keras는 4줄 처리로 끝남
- sigmoid와, binary_crossentropy만 넣어주면 됨
- 이 내부 구조를 알 수는 없다
- 내부 구조를 알기 위해서는 tensorflow를 사용해야 함
